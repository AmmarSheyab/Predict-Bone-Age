# train_test_split data

```python
from sklearn.model_selection import train_test_split
df_train, df_valid = train_test_split(train_csv, test_size = 0.10, random_state = 42)
```

```python

# Data Preprocessing

#horizontal_flip : flip inputs horizontally

#width_shift_range : shift the image to the left or right

#height_shift_range :shift vertically(up or down)

#rotation_range : The degree of rotation of the image is random

#zoom_range : Range for random zoom
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.25,
                                   rotation_range = 5,
                                   height_shift_range = 0.15, 
                                   width_shift_range = 0.15,
                                   horizontal_flip=True,
                                   )

val_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.25,
                                   height_shift_range = 0.15, 
                                   width_shift_range = 0,
                                   horizontal_flip=True,
                                   rotation_range = 5,
                                   )


test_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.25,
                                   height_shift_range = 0.15, 
                                   width_shift_range = 0,
                                   horizontal_flip=True,
                                   rotation_range = 5,
                                   )

```

# Creating the train set

```python
data_file_train=train_datagen.flow_from_dataframe(dataframe=df_train,
                                                  directory=(r'C:\Users\ASUS\Desktop\fainal project\image\train1\train'),
                                                  x_col='images',
                                                  y_col='bone_age_z',
                                                  target_size=(385,385),
                                                  batch_size=32,
                                                   color_mode = 'rgb',
                                                  class_mode='other',
                                                  
                                                )




data_file_val=train_datagen.flow_from_dataframe(dataframe=df_valid,
                                                  directory=(r'C:\Users\ASUS\Desktop\fainal project\image\train1\train'),
                                                  x_col='images',
                                                  y_col='bone_age_z',
                                                  target_size=(385, 385),
                                                  batch_size=32,
                                                 color_mode = 'rgb',
                                                  class_mode='other',
                                                )

test_X, test_Y=next((test_datagen.flow_from_dataframe(dataframe=df_valid,
                                                  directory=(r'C:\Users\ASUS\Desktop\fainal project\image\train1\train'),
                                                  x_col='images',
                                                  y_col='bone_age_z',
                                                  target_size=(385, 385),
                                                  batch_size=1262,
                                                       color_mode = 'rgb',
                                                  class_mode='other',
                                                )))
 
```

# function to return mae in months

```python
from keras.metrics import mean_absolute_error
def mae_in_months(x_p, y_p):
    return mean_absolute_error((std_bone_age*x_p + mean_bone_age), (std_bone_age*y_p + mean_bone_age))
```

# Building the CNN

```python
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping,ReduceLROnPlateau
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Dropout
cnn = tf.keras.models.Sequential()
# Convolution
cnn.add(tf.keras.layers.Conv2D(filters=16, kernel_size=(3,3), padding="same", activation="relu", input_shape=(385, 385, 3)))
#maxpooling
cnn.add(tf.keras.layers.MaxPool2D(2,2))
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), padding="same", activation="relu",input_shape=(385, 385, 3)))
cnn.add(tf.keras.layers.MaxPool2D(2,2))
cnn.add(Dropout(0.2))
cnn.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu",input_shape=(385, 385, 3)))
cnn.add(tf.keras.layers.MaxPool2D(2,2))
cnn.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu",input_shape=(385, 385, 3)))
cnn.add(tf.keras.layers.MaxPool2D(2,2))
cnn.add(Dropout(0.2))
cnn.add(tf.keras.layers.Flatten())
# Full Connection
cnn.add(tf.keras.layers.Dense(units=64, activation='relu'))
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
cnn.add(Dropout(0.3))
output=cnn.add(tf.keras.layers.Dense(units=1, activation='linear'))
cnn.compile(loss ='mse', optimizer= 'adam', metrics = [mae_in_months] )
cnn.summary()
```

```python
#early stopping
early_stopping = EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience= 5,
                              verbose=0, mode='auto')
```

```python
#model checkpoint
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', save_best_only=True)
```

```python
#reduce lr on plateau
red_lr_plat = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=10, verbose=0, mode='auto', min_delta=0.0001, cooldown=0, min_lr=0)
```

```python
callbacks = [early_stopping,mc, red_lr_plat]
```

# fit model

```python
history=cnn.fit_generator(data_file_train,
                  steps_per_epoch = 315,
                  epochs = 50,
                  validation_data = (data_file_val),
                  validation_steps=1,
                  callbacks= callbacks)

  
history                        
```

```python
#cnn.load_weights('ammar.h5')
pred = mean_bone_age + std_bone_age*(cnn.predict(test_X, batch_size = 32, verbose = True))
test_months = mean_bone_age + std_bone_age*(test_Y)

ord_ind = np.argsort(test_Y)
ord_ind = ord_ind[np.linspace(0, len(ord_ind)-1, 8).astype(int)] # take 8 evenly spaced ones
fig, axs = plt.subplots(4, 2, figsize = (15, 30))
for (ind, ax) in zip(ord_ind, axs.flatten()):
    ax.imshow(test_X[ind, :,:,0], cmap = 'bone')
    ax.set_title('Age: %fY\nPredicted Age: %fY' % (test_months[ind]/12.0, 
                                                           pred[ind]/12.0))
    ax.axis('off')
fig.savefig('trained_image_predictions.png', dpi = 300)
```
